```python
# Her kursun seviyesine ,Kursun konusuna ,Ödenmiş kurs ve yıla göre maksimum obone sayısı :
def ObuneSayisiMX(sayiX):
    for t in sayiX:
        print(f"========================== {t} ' a  göre maksimum obone sayısı \n")
        print(df.groupby(t)['Obune sayısı'].max())
        print("====================================================================")
        
ListObune=['Kursun konusu','Seviye','Yıl','Ödenmiş kurs']
ObuneSayisiMX(ListObune)
```

```python
# Her kursun seviyesine ,Kursun konusuna ,'Ödenmiş kurs ve yıla göre Minimum obone sayısı :
def ObuneSayisiMn(sayiN):
    for t in sayiN:
        print(f"========================== {t} ' a  göre minimum obone sayısı \n")
        print(df.groupby(t)['Obune sayısı'].min())
        print("====================================================================")
        
ListObune=['Kursun konusu','Seviye','Yıl','Ödenmiş kurs']
ObuneSayisiMn(ListObune)
```
