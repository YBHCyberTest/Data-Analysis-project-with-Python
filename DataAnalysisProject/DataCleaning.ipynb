```python
#course_id sütunu silme :
df.drop("course_id",axis=1,inplace=True)
```

```python
#is_paid sütunu Ture ,false değerlerinden Yes ,No değerlerine değiştirme :
#df['is_paid'] = df['is_paid'].astype('object')
#df['is_paid'] = df['is_paid'].replace(['True'], pice sütunun tipi object'ten int64'e değiştirme ve free kalemesi yerinde 0 değeri gösterme:'yer')
```

```python
df['price'] = df['price'].replace(['Free'],0)
df['price'] = df['price'].astype('int64')
```

```python
df['content_duration'] = [x.replace('hours', '').replace('hour', '').replace('mins', '').replace('questions', '')for x in df['content_duration']]
df['content_duration'] = df['content_duration'].astype(str).astype(float)
```

```python
df['published_timestamp']= pd.to_datetime(df['published_timestamp'])
```

```python
df.columns=['Kurs başlığı','Ödenmiş kurs','Fıyat','Obune sayısı','Görüntülme Sayısı','Ders Sayısı','Seviye','İçerik Süresi','yayınlanan zaman damgası','Kursun konusu']
```
